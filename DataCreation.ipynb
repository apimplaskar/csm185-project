```python
import pandas as pd
import numpy as np
```

```python
chartevents = pd.read_csv("chartEventsFiltered.csv")
labevents = pd.read_csv("labEventsFiltered.csv")
feb18data = pd.read_csv("Feb18-Data.csv", low_memory=False)
feb18data.head()

#reading in tables
```

```python
feb18data = feb18data.loc[feb18data['ITEMID'].isin(["224284","224285","224287","224288","224290","226724","226515"])]

# filtering out unnecessary ITEMIDs
#224284 - Arterial line Cap Change
#224285 - Arterial line Change over Wire Date
#224287 - Arterial Line Dressing Change
#224288 - Arterial line Insertion Date
#224290 - Arterial line Tubing Change
#226724 - Hospital Admit Date
#226515 - Date of Birth
```

```python
feb18data = feb18data.set_index(["ROW_ID", "SUBJECT_ID", "HADM_ID", "CHARTTIME", "VALUEUOM", "diagnosis", "procedure", "deceased", "ADMITTIME", "DISCHTIME", "DEATHTIME", "ETHNICITY", "HOSPITAL_EXPIRE_FLAG", "HAS_CHARTEVENTS_DATA", "ITEMID"])["VALUE"].unstack()

# "unmelting" ITEMID column; creating a row for every unique value in ITEMID
```

```python
feb18data = feb18data.rename_axis(columns = None).reset_index()

#removing ITEMID label and resetting index
```

```python
feb18data = feb18data.rename(columns={"diagnosis": "DIAGNOSIS", "procedure": "PROCEDURE", "deceased": "DECEASED", 224284: "Arterial line Cap Change", 224285: "Arterial line Change over Wire Date", 224287: "Arterial Line Dressing Change", 224288: "Arterial line Insertion Date", 224290: "Arterial line Tubing Change", 226515: "Date of Birth", 226724: "Hospital Admit Date"})
feb18data.head()

#renaming columns (including ITEMIDs)
```

```python
labevents.head()
```

```python
labevents = labevents.drop(columns=["valuenum", "flag"])
labevents.columns = ["ROW_ID", "SUBJECT_ID", "HADM_ID", "ITEMID", "CHARTTIME", "VALUE", "VALUEUOM"]

# dropping 2 columns and renaming the rest for uniformity
```

```python
labevents = labevents.set_index(["ROW_ID", "SUBJECT_ID", "HADM_ID", "CHARTTIME", "VALUEUOM", "ITEMID"])["VALUE"].unstack()

# unmelting ITEMID column
```

```python
labevents = labevents.rename_axis(columns = None).reset_index()

#removing ITEMID label and resetting index
```

```python
labevents = labevents.rename(columns={50817: "Oxygen Saturation", 50893: "Calcium, Total", 50904: "HDL", 50906: "LDL", 50907: "Cholesterol", 50912: "Creatinine", 51006: "Urea Nitrogen", 51108: "Urine Volume", 51237: "INR(PT)", 51265: "Platelet Count", 51279: "Red Blood Cells", 51300: "WBC Count"})

#renaming columns
```

```python
feb18data = pd.merge(feb18data, labevents, how='left', on=['SUBJECT_ID', 'ROW_ID', "HADM_ID", "CHARTTIME", "VALUEUOM"], right_index=True, left_index=True)

# left join of feb18data on labevents table 
```

```python
chartevents = chartevents.rename_axis(columns = None).reset_index()

#removing ITEMID label and resetting index
```

```python
feb18data = pd.merge(feb18data, chartevents, how='inner', on=['SUBJECT_ID', 'ROW_ID', "HADM_ID", "CHARTTIME", "VALUEUOM"], right_index=True, left_index=True)

# inner join of feb18data and processed chartevents table 
```
